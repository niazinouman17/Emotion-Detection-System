```python
import os
import librosa
import numpy as np
import soundfile as sf

# Define constants
INPUT_DIR =   "C:\\Users\\New Classic\\Desktop\\AI and ML\\Data Set\\Neutral"
 # Existing dataset folder
OUTPUT_DIR_2S = "processed_2_seconds_Main"  # New folder for 2-second clips
OUTPUT_DIR_3S = "processed_3_seconds_Main"  # New folder for 3-second clips
TARGET_SR = 22050 # Target sample rate
DURATION_2S = 2  # Target duration in seconds for 2-second clips
DURATION_3S = 3  # Target duration in seconds for 3-second clips

# Create output directories if they don't exist
os.makedirs(OUTPUT_DIR_2S, exist_ok=True)
os.makedirs(OUTPUT_DIR_3S, exist_ok=True)

def process_audio(file_path, target_duration, sr, mode="truncate"):
    """Process audio: truncate or pad to a target duration."""
    y, _ = librosa.load(file_path, sr=sr)
    target_samples = target_duration * sr
    if len(y) > target_samples:
        return y[:target_samples]  # Truncate
    elif len(y) < target_samples:
        return np.pad(y, (0, target_samples - len(y)), mode='constant')  # Pad
    return y  # Return as-is if already matches the target duration

def process_dataset(input_dir, output_dir_2s, output_dir_3s, sr):
    """Process the dataset into 2-second and 3-second clips."""
    for root, _, files in os.walk(input_dir):
        for file in files:
            if file.endswith(".wav"):  # Process only .wav files
                file_path = os.path.join(root, file)
                
                # Process for 2 seconds
                y_2s = process_audio(file_path, DURATION_2S, sr, mode="truncate")
                output_path_2s = os.path.join(output_dir_2s, file)
                sf.write(output_path_2s, y_2s, sr)
                
                # Process for 3 seconds
                y_3s = process_audio(file_path, DURATION_3S, sr, mode="pad")
                output_path_3s = os.path.join(output_dir_3s, file)
                sf.write(output_path_3s, y_3s, sr)
                print(f"Processed: {file}")

# Run the processing
process_dataset(INPUT_DIR, OUTPUT_DIR_2S, OUTPUT_DIR_3S, TARGET_SR)
print("Processing complete. Check the 'processed_2_seconds' and 'processed_3_seconds' directories.")
```

```python
import librosa

def feature_chromagram(waveform, sample_rate):
    # STFT computed here explicitly; mel spectrogram and MFCC functions do this under the hood
    stft_spectrogram=np.abs(librosa.stft(waveform))
    # Produce the chromagram for all STFT frames and get the mean of each column of the resulting matrix to create a feature array
    chromagram=np.mean(librosa.feature.chroma_stft(S=stft_spectrogram, sr=sample_rate).T,axis=0)
    return chromagram

def feature_melspectrogram(waveform, sample_rate):
    # Produce the mel spectrogram for all STFT frames and get the mean of each column of the resulting matrix to create a feature array
    # Using 8khz as upper frequency bound should be enough for most speech classification tasks
    melspectrogram=np.mean(librosa.feature.melspectrogram(y=waveform, sr=sample_rate, n_mels=128, fmax=8000).T,axis=0)
    return melspectrogram

def feature_mfcc(waveform, sample_rate):
    # Compute the MFCCs for all STFT frames and get the mean of each column of the resulting matrix to create a feature array
    # 40 filterbanks = 40 coefficients
    mfc_coefficients=np.mean(librosa.feature.mfcc(y=waveform, sr=sample_rate, n_mfcc=40).T, axis=0) 
    return mfc_coefficients
```

```python
def get_features(file):
    # load an individual soundfile
     with soundfile.SoundFile(file) as audio:
        waveform = audio.read(dtype="float32")
        sample_rate = audio.samplerate
        # compute features of soundfile
        chromagram = feature_chromagram(waveform, sample_rate)
        melspectrogram = feature_melspectrogram(waveform, sample_rate)
        mfc_coefficients = feature_mfcc(waveform, sample_rate)

        feature_matrix=np.array([])
        # use np.hstack to stack our feature arrays horizontally to create a feature matrix
        feature_matrix = np.hstack((chromagram, melspectrogram, mfc_coefficients))
        
        return feature_matrix
```

```python
emotions = {
    '01':'Angry',
    '02':'Happy',
    '03':'Neutral',
    '04':'Sad'
}
```

```python
import os, glob

def load_data():
    X,y=[],[]
    count = 0
    for file in glob.glob("C:\\xampp\\htdocs\\FYP\\Emotion-Detection-System\\processed_3_seconds_Main\\*.wav"):
        file_name=os.path.basename(file)
        emotion=emotions[file_name.split("-")[2]]
        features = get_features(file)
        X.append(features)
        y.append(emotion)
        count += 1
        # '\r' + end='' results in printing over same line
        print('\r' + f' Processed {count}/{200} audio samples',end=' ')
    # Return arrays to plug into sklearn's cross-validation algorithms
    return np.array(X), np.array(y)
```

```python
features, emotions = load_data()
```
