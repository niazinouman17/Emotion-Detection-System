```python
import os
import librosa
import numpy as np
import soundfile as sf

# Define constants
INPUT_DIR =   "C:\\Users\\New Classic\\Desktop\\AI and ML\\Data Set\\Neutral"
 # Existing dataset folder
OUTPUT_DIR_2S = "processed_2_seconds_Main"  # New folder for 2-second clips
OUTPUT_DIR_3S = "processed_3_seconds_Main"  # New folder for 3-second clips
TARGET_SR = 22050 # Target sample rate
DURATION_2S = 2  # Target duration in seconds for 2-second clips
DURATION_3S = 3  # Target duration in seconds for 3-second clips

# Create output directories if they don't exist
os.makedirs(OUTPUT_DIR_2S, exist_ok=True)
os.makedirs(OUTPUT_DIR_3S, exist_ok=True)

def process_audio(file_path, target_duration, sr, mode="truncate"):
    """Process audio: truncate or pad to a target duration."""
    y, _ = librosa.load(file_path, sr=sr)
    target_samples = target_duration * sr
    if len(y) > target_samples:
        return y[:target_samples]  # Truncate
    elif len(y) < target_samples:
        return np.pad(y, (0, target_samples - len(y)), mode='constant')  # Pad
    return y  # Return as-is if already matches the target duration

def process_dataset(input_dir, output_dir_2s, output_dir_3s, sr):
    """Process the dataset into 2-second and 3-second clips."""
    for root, _, files in os.walk(input_dir):
        for file in files:
            if file.endswith(".wav"):  # Process only .wav files
                file_path = os.path.join(root, file)
                
                # Process for 2 seconds
                y_2s = process_audio(file_path, DURATION_2S, sr, mode="truncate")
                output_path_2s = os.path.join(output_dir_2s, file)
                sf.write(output_path_2s, y_2s, sr)
                
                # Process for 3 seconds
                y_3s = process_audio(file_path, DURATION_3S, sr, mode="pad")
                output_path_3s = os.path.join(output_dir_3s, file)
                sf.write(output_path_3s, y_3s, sr)
                print(f"Processed: {file}")

# Run the processing
process_dataset(INPUT_DIR, OUTPUT_DIR_2S, OUTPUT_DIR_3S, TARGET_SR)
print("Processing complete. Check the 'processed_2_seconds' and 'processed_3_seconds' directories.")
```

```python
with soundfile.SoundFile( "C:\\xampp\\htdocs\\FYP\\Emotion-Detection-System\\processed_2_seconds_Main\\Happy\\SF1_F1_H01.wav"
) as audio:
    happy_waveform = audio.read(dtype="float32")
    sample_rate = audio.samplerate

with soundfile.SoundFile("C:\\xampp\\htdocs\\FYP\\Emotion-Detection-System\\processed_2_seconds_Main\\Angry\\SM1_F1_A01.wav") as audio:
    angry_waveform = audio.read(dtype="float32")
    # same sample rate
```

```python
stft_spectrum_matrix = librosa.stft(happy_waveform)
plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.amplitude_to_db(np.abs(stft_spectrum_matrix), ref=np.max),y_axis='log', x_axis='time')
plt.title('Happy STFT Power spectrogram')
plt.colorbar(format='%+2.0f dB')
plt.tight_layout()

stft_spectrum_matrix = librosa.stft(angry_waveform)
plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.amplitude_to_db(np.abs(stft_spectrum_matrix), ref=np.max),y_axis='log', x_axis='time')
plt.title('Angry STFT Power spectrogram')
plt.colorbar(format='%+2.0f dB')
plt.tight_layout()
```

```python
from matplotlib.colors import Normalize
mfc_coefficients = librosa.feature.mfcc(y=happy_waveform, sr=sample_rate, n_mfcc=40)
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfc_coefficients, x_axis='time',norm=Normalize(vmin=-30,vmax=30))
plt.colorbar()
plt.yticks(())
plt.ylabel('MFC Coefficient')
plt.title('Happy MFC Coefficients')
plt.tight_layout()
                                        
mfc_coefficients = librosa.feature.mfcc(y=angry_waveform, sr=sample_rate, n_mfcc=40)
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfc_coefficients, x_axis='time',norm=Normalize(vmin=-30,vmax=30))
plt.colorbar()
plt.yticks(())
plt.ylabel('MFC Coefficient')
plt.title('Angry MFC Coefficients')
plt.tight_layout()
```

```python
melspectrogram = librosa.feature.melspectrogram(y=happy_waveform, sr=sample_rate, n_mels=128, fmax=8000)
plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.power_to_db(S=melspectrogram, ref=np.mean),y_axis='mel',fmax=8000, x_axis='time', norm=Normalize(vmin=-20,vmax=20))
plt.colorbar(format='%+2.0f dB',label='Amplitude')
plt.ylabel('Mels')
plt.title('Happy Mel spectrogram')
plt.tight_layout()

melspectrogram = librosa.feature.melspectrogram(y=angry_waveform, sr=sample_rate, n_mels=128, fmax=8000)
plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.power_to_db(S=melspectrogram, ref=np.mean),y_axis='mel',fmax=8000, x_axis='time', norm=Normalize(vmin=-20,vmax=20))
plt.colorbar(format='%+2.0f dB',label='Amplitude')
plt.ylabel('Mels')
plt.title('Angry Mel spectrogram')
plt.tight_layout()
```
